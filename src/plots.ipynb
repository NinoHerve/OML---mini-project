```python
# Standard library imports
import os

# Third party imports
import matplotlib.pyplot as plt
import pandas as pd
```

```python
metric_folder = "metrics/lr_range_test"
file_names = ["CIFAR10_rangeLR.csv", "FashionMNIST_rangeLR.csv"]  # Add your file names here

plt.figure(figsize=(6, 3))

for file_name in file_names:
    df = pd.read_csv(os.path.join(metric_folder, file_name))
    lrs = df[df["source"] == "test"]["lr"].to_numpy()
    accuracy = df[df["source"] == "test"]["accuracy"].to_numpy()
    
    plt.plot(lrs, accuracy, label=file_name.split(".")[0])

plt.legend()
plt.xlabel("Learning Rate")
plt.ylabel("Top-1 Accuracy")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
metric_folder = "metrics/lr_range_pretrain_test"
file_names = ["CIFAR10_rangeLR_pretrained.csv", "FashionMNIST_rangeLR_pretrained.csv"]  # Add your file names here

plt.figure(figsize=(6, 3))

for file_name in file_names:
    df = pd.read_csv(os.path.join(metric_folder, file_name))
    lrs = df[df["source"] == "test"]["lr"].to_numpy()
    accuracy = df[df["source"] == "test"]["accuracy"].to_numpy()
    
    plt.plot(lrs, accuracy, label=file_name.split(".")[0][:-11])

plt.legend()
plt.xlabel("Learning Rate")
plt.ylabel("Top-1 Accuracy")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
metric_folder = "metrics/FashionMNIST_lr_schedulers"
file_names = ["FashionMNIST_FixLR.csv", "FashionMNIST_LinearLR.csv", "FashionMNIST_OneCycleLR.csv", "FashionMNIST_CyclicLR.csv", "FashionMNIST_CyclicLR2.csv"]  # Add your file names here

plt.figure(figsize=(6, 3))

for file_name in file_names:
    df = pd.read_csv(os.path.join(metric_folder, file_name))
    lrs = df[df["source"] == "test"]["iter"].to_numpy()
    accuracy = df[df["source"] == "test"]["lr"].to_numpy()
    
    plt.plot(lrs, accuracy, label=file_name.split(".")[0].split("_")[1])

plt.legend()
plt.xlabel("Training Steps")
plt.ylabel("Learning Rate")
plt.grid()
plt.tight_layout()
plt.show()
```
